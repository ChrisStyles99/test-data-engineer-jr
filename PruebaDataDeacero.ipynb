```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 2000)
pd.set_option('display.max_columns', 500)
```

```python
pasajeros2016 = pd.read_json('http://analytics.deacero.com/Api/GetApi/ApiPasajeros2016/ecfb5fc7-0932-590f-832c-6d6055f2be07')
pasajeros2017 = pd.read_json('http://analytics.deacero.com/Api/GetApi/ApiPasajeros2017/faabd632-cc39-552d-a68b-02de4242f636').drop_duplicates()
```

```python
pasajeros2016
```

```python
pasajeros2017
```

```python
pasajeros = pasajeros2016.append(pasajeros2017, ignore_index=True)
pasajeros
```

```python
vuelos2016 = pd.read_json('http://analytics.deacero.com/Api/GetApi/ApiVuelos2016/9ea3b836-6938-52dc-9626-a8e35db81dd5')
vuelos2017 = pd.read_json('http://analytics.deacero.com/Api/GetApi/ApiVuelos2017/fc126260-1cf8-5a46-995d-ba639ff5868b')
```

```python
vuelos2016
```

```python
vuelos2017
```

```python
vuelos = vuelos2016.append(vuelos2017, ignore_index=True)
vuelos
```

```python
pasajeros.columns
```

```python
vuelos.columns
```

```python
pasajeros_vuelos = pd.merge(vuelos, pasajeros, left_on='Cve_Cliente', right_on='ID_Pasajero')
pasajeros_vuelos
```

```python
lineas = pd.read_json('http://analytics.deacero.com/Api/GetApi/ApiLineaAerea/1a8d9e13-ce30-50fc-bf34-6490eb799a75')
lineas
```

```python
pasajeros_vuelos_lineas = pd.merge(lineas, pasajeros_vuelos, how='outer', left_on='Code', right_on='Cve_LA').replace(np.nan, 'Otra')
pasajeros_vuelos_lineas[['Linea_Aerea', 'Viaje', 'Clase', 'Precio', 'Ruta', 'Edad']]
```

```python
pasajeros_vuelos_lineas['Viaje'] = pd.to_datetime(pasajeros_vuelos_lineas['Viaje'])
```

```python
pasajeros_vuelos_lineas[['Linea_Aerea', 'Viaje', 'Clase', 'Precio', 'Ruta', 'Edad']]
```

```python
pasajeros_vuelos_lineas[(pasajeros_vuelos_lineas['Viaje'] > pd.Timestamp(2016,1,1)) & (pasajeros_vuelos_lineas['Viaje'] < pd.Timestamp(2016,6,30))]
```

```python
ene_jun2016 = pasajeros_vuelos_lineas[(pasajeros_vuelos_lineas['Viaje'] > pd.Timestamp(2016,1,1)) & (pasajeros_vuelos_lineas['Viaje'] < pd.Timestamp(2016,6,30))]
ene_jun2016
```

```python
economy_ene_jun2016 = ene_jun2016[ene_jun2016['Clase'] == 'Economy']
economy_ene_jun2016
```

```python
for x in economy_ene_jun2016['Ruta'].drop_duplicates():
    for y in economy_ene_jun2016[economy_ene_jun2016['Ruta'] == x]['Linea_Aerea'].drop_duplicates():
        print('Economy Ene-Jun 2016', x, y)
        print(economy_ene_jun2016[(economy_ene_jun2016['Linea_Aerea'] == y) & (economy_ene_jun2016['Ruta'] == x)]['Precio'].mean())
```

```python
business_ene_jun2016 = ene_jun2016[ene_jun2016['Clase'] == 'Business']
business_ene_jun2016
```

```python
for x in business_ene_jun2016['Ruta'].drop_duplicates():
    for y in business_ene_jun2016[business_ene_jun2016['Ruta'] == x]['Linea_Aerea'].drop_duplicates():
        print('Business Ene-Jun 2016', x, y)
        print(business_ene_jun2016[(business_ene_jun2016['Linea_Aerea'] == y) & (business_ene_jun2016['Ruta'] == x)]['Precio'].mean())
```

```python
first_class_ene_jun2016 = ene_jun2016[ene_jun2016['Clase'] == 'First Class']
first_class_ene_jun2016
```

```python
for x in first_class_ene_jun2016['Ruta'].drop_duplicates():
    for y in first_class_ene_jun2016[first_class_ene_jun2016['Ruta'] == x]['Linea_Aerea'].drop_duplicates():
        print('First Class Ene-Jun 2016', x, y)
        print(first_class_ene_jun2016[(first_class_ene_jun2016['Linea_Aerea'] == y) & (first_class_ene_jun2016['Ruta'] == x)]['Precio'].mean())
```
