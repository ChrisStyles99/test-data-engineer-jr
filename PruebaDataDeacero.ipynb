```python
pasajeros_vuelos = pd.merge(vuelos, pasajeros, left_on='Cve_Cliente', right_on='ID_Pasajero')
pasajeros_vuelos
```

```python
lineas = pd.read_json('http://analytics.deacero.com/Api/GetApi/ApiLineaAerea/1a8d9e13-ce30-50fc-bf34-6490eb799a75')
lineas
```

```python
pasajeros_vuelos_lineas = pd.merge(lineas, pasajeros_vuelos, how='outer', left_on='Code', right_on='Cve_LA').replace(np.nan, 'Otra')
pasajeros_vuelos_lineas[['Linea_Aerea', 'Viaje', 'Clase', 'Precio', 'Ruta', 'Edad']]
```

```python
pasajeros_vuelos_lineas['Viaje'] = pd.to_datetime(pasajeros_vuelos_lineas['Viaje'])
```

```python
ene_jun2016 = pasajeros_vuelos_lineas[(pasajeros_vuelos_lineas['Viaje'] > pd.Timestamp(2016,1,1)) & (pasajeros_vuelos_lineas['Viaje'] < pd.Timestamp(2016,6,30))]
ene_jun2016
```

```python
jul_dic2016 = pasajeros_vuelos_lineas[(pasajeros_vuelos_lineas['Viaje'] > pd.Timestamp(2016,7,1)) & (pasajeros_vuelos_lineas['Viaje'] < pd.Timestamp(2016,12,31))]
jul_dic2016
```

```python
ene_jun2017 = pasajeros_vuelos_lineas[(pasajeros_vuelos_lineas['Viaje'] > pd.Timestamp(2017,1,1)) & (pasajeros_vuelos_lineas['Viaje'] < pd.Timestamp(2017,6,30))]
ene_jun2017
```

```python
jul_dic2017 = pasajeros_vuelos_lineas[(pasajeros_vuelos_lineas['Viaje'] > pd.Timestamp(2017,7,1)) & (pasajeros_vuelos_lineas['Viaje'] < pd.Timestamp(2017,12,31))]
jul_dic2017
```

```python
def makerows(dataFrame):
    newlist = list()
    promedios = list()
    for a in dataFrame['Clase'].drop_duplicates():
        for x in dataFrame[dataFrame['Clase'] == a]['Ruta'].drop_duplicates():
            for y in dataFrame[dataFrame['Ruta'] == x]['Linea_Aerea'].drop_duplicates():
                newlist.append(dataFrame[(dataFrame['Linea_Aerea'] == y) & (dataFrame['Ruta'] == x) & (dataFrame['Clase'] == a)][['Clase', 'Ruta', 'Linea_Aerea']])
                promedios.append(dataFrame[(dataFrame['Linea_Aerea'] == y) & (dataFrame['Ruta'] == x) & (dataFrame['Clase'] == a)]['Precio'].mean())
    df = pd.concat(newlist, ignore_index=True)
    df['Promedios'] = pd.Series(promedios)
    df['Promedios'] = df['Promedios'].replace(np.nan, 'No info')
    return df
```

```python
dfej2016 = makerows(ene_jun2016)
dfej2016
```

```python
dfjd2016 = makerows(jul_dic2016)
dfjd2016
```

```python
dfej2017 = makerows(ene_jun2017)
dfej2017
```

```python
dfjd2017 = makerows(jul_dic2017)
dfjd2017
```
